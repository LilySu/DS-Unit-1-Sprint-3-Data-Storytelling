## Reproduce with real data

```python
df = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/inconvenient-sequel/ratings.csv')
```

```python
df.shape
```

```python
df.head(3)
```

```python
pd.set_option('display.max_columns',500)
df.head(3)
```

```python
df.T
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
df['timestamp'].describe()#all unique, once we set index, we can slice dataframe to date
```

```python
df.set_index('timestamp', inplace=True)
```

```python
df['2017-08-29'].head(3)
```

```python
df['category'].value_counts()#each category is present in each time of observation
```

```python
lastday =df['2017-08-29']
```

```python
lastday[lastday['category']=='IMDb users'].respondents.plot()
```

```python
final = lastday[lastday['category']=='IMDb users'].tail(1)
final
```

```python
pct_columns = ['1_pct', '2_pct', '3_pct', '4_pct', '5_pct',
               '6_pct', '7_pct', '8_pct', '9_pct', '10_pct']
final[pct_columns]
```

```python
data = final[pct_columns].T
data.index = range(1, 11)
data
```

```python
plt.style.use('fivethirtyeight')
fig = plt.figure();
fig.patch.set_facecolor('#f9f9f9');

ax = data.plot.bar(color='#ef7030', width=0.9)

ax.set(yticks=range(-3, 50, 10),
       facecolor='#f9f9f9');
#ax.set_axis_bgcolor('#f9f9f9')
#ax.set_facecolor('white')

#plt.title("'An Inconvenient Sequel: Truth to Power' is divisive")
#plt.ylabel('Percent of total votes')
#plt.xlabel('Rating')
plt.ylabel('Percent of Total Votes', fontsize=10, fontweight='bold');
plt.xlabel('Rating', fontsize=10, fontweight='bold');
ax.get_legend().remove()

ax.text(x=-2, y=55, s="'An Inconvenient Sequel: Truth to Power' is divisive",
       fontsize=14, fontweight='bold');
ax.text(x=-2, y=52, s="IMDB ratings for the film as of Aug. 29",
       fontsize=12);
ax.tick_params(labelrotation=0);
```

```python
import altair as alt
from vega_datasets import data

source = data.movies.url

alt.Chart(source).mark_bar().encode(
    alt.X("IMDB_Rating:Q", bin=True),
    y='count()',
)
```

```python
import altair as alt
from vega_datasets import data

source=data.cars()

alt.Chart(source).mark_circle().encode(
    alt.X(alt.repeat("column"), type='quantitative'),
    alt.Y(alt.repeat("row"), type='quantitative'),
    color='Origin:N'
).properties(
    width=150,
    height=150
).repeat(
    row=['Horsepower', 'Acceleration', 'Miles_per_Gallon'],
    column=['Miles_per_Gallon', 'Acceleration', 'Horsepower']
).interactive()
```
